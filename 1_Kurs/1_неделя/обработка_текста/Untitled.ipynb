```python
with open("sentences.txt") as file:
    data = list(file)
```

```python
datal = []
for line in data:
    datal.append(line.lower())
datal
```

```python
import re
dataw = []
for line in datal:
    dataw.append(list(filter(lambda c: c!= '', re.split('[^a-z]', line))))
for line in dataw:
    print(line)
```

```python
dictw = {}

i = 0
for line in dataw:
    for word in line:
        if word not in dictw:
            dictw[word] = i
            i+=1
```

```python
dictw.keys()
```

```python
import pandas as pd
len(dictw)
```

```python
dataframe = pd.DataFrame(dictw, range(len(dataw)))
```

```python
a, b = dataframe.shape
```

```python
for k in range(a):
    for j in range(b):
        dataframe.iloc[k,j] = 0
dataframe
```

```python
for i in range(len(dataw)):
    for word in dataw[i]:
        dataframe.loc[dataframe.index[i], word] +=1
dataframe
```

```python
from scipy.spatial.distance import cosine
```

```python
distanceFromFirstSentence = []

for i in range(22):
    distanceFromFirstSentence.append(cosine(dataframe.iloc[0], dataframe.iloc[i]))
print(distanceFromFirstSentence)
```

```python
distanceFromFirstSentenceCopy = list(distanceFromFirstSentence)
twoClosestValues = [[-1, 0], [-1, 0]]

distanceFromFirstSentenceCopy.remove(min(distanceFromFirstSentenceCopy))


for n in range(2):
    twoClosestValues[n][1] = min(distanceFromFirstSentenceCopy)
    for j in range(len(distanceFromFirstSentence)):
        if twoClosestValues[n][1] == distanceFromFirstSentence[j]:
            twoClosestValues[n][0] = j
            distanceFromFirstSentenceCopy.remove(min(distanceFromFirstSentenceCopy))
            
print(twoClosestValues)
```

```python
with open('answer.txt', 'w') as file:
    for k in range(2):
        file.write(str(twoClosestValues[k][0]) + ' ')
```
